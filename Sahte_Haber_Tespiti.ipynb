## Metin ön işleme

#### Metinleri küçük harfe dönüştürme, bağlaçları kaldırma, noktalama işaretlerini kaldırma, linkleri kaldırma işlemi vb

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+',b'',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
import pandas as pd
import numpy as np
import networkx as nx
from sklearn.feature_extraction.text import CountVectorizer
```

Malatya Centrality

```python
# Haber metinlerini kelime vektörlerine dönüştür
vectorizer = CountVectorizer(stop_words='english', max_features=500)  # En sık geçen 5000 kelimeyi kullan
X = vectorizer.fit_transform(data['text']).toarray()
feature_names = vectorizer.get_feature_names_out()

# Özelliklerin ve verinin şekline bak
print(f"Veri şekli: {X.shape}")  # Örnek: (20000, 5000)
```

```python
# Graf oluştur
graph = nx.Graph()

# Her haberi bir düğüm olarak ekle
for i in range(len(data)):
    graph.add_node(i, text=data['text'][i], label=data['class'][i])

# Kenarları ve ağırlıkları ekle
for i in range(len(data)):
    for j in range(i + 1, len(data)):
        # Ortak kelime sayısını hesapla
        common_words = np.sum(X[i] * X[j])
        if common_words > 0:
            graph.add_edge(i, j, weight=common_words)
```

```python
# Düğüm derecelerini hesapla
degree_dict = {}
for node in graph.nodes():
    # Düğümün derecesi: bağlı kenarların ağırlıklarının toplamı
    degree_dict[node] = sum(edge[2]['weight'] for edge in graph.edges(node, data=True))

# Komşuların derecelerini hesapla
neighbor_degree_dict = {}
for node in graph.nodes():
    neighbor_degrees = sum(degree_dict[neighbor] for neighbor in graph.neighbors(node))
    neighbor_degree_dict[node] = neighbor_degrees
```

```python
# Dereceye dayalı skor hesapla
score_dict = {}
for node in graph.nodes():
    if neighbor_degree_dict[node] > 0:
        score_dict[node] = degree_dict[node] / neighbor_degree_dict[node]
    else:
        score_dict[node] = 0  # Komşusu yoksa skor 0
```

```python
# Skorlara göre sahte haber tespiti
threshold = 0.5  # Eşik değeri (optimize edilebilir)
predictions = {node: (1 if score > threshold else 0) for node, score in score_dict.items()}

# Sonuçları bir DataFrame'e yaz
results = pd.DataFrame({
    'text': [graph.nodes[node]['text'] for node in graph.nodes()],
    'actual_label': [graph.nodes[node]['label'] for node in graph.nodes()],
    'predicted_label': [predictions[node] for node in graph.nodes()],
    'score': [score_dict[node] for node in graph.nodes()]
})

# İlk birkaç sonucu gör
print(results.head())
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Gerçek ve tahmin edilen etiketleri al
y_true = results['actual_label']
y_pred = results['predicted_label']

# Doğruluk ve diğer metrikler
print("Accuracy:", accuracy_score(y_true, y_pred))
print("\nClassification Report:\n", classification_report(y_true, y_pred))
```
